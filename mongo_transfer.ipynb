```python
#Load our csv file path
csvfile = open('Datasets/cleaned_data.csv', 'r')
```

```python
features = ["Gender","Age","Caffeine (mg)","Alcohol (gm)","Weight (kg)","Standing Height (cm)","Pulse regular or irregular","Systolic: Blood pressure mm Hg","Direct HDL-Cholesterol (mg/dL)","Avg Drinks per Day","Smoker?"]

reader = csv.DictReader(csvfile)

header = features

for each in reader:
    row = {}
    for field in header:
        row[field] = each[field]
#person_features is a collection within the database project-2
    db.person_features.insert_one(row)
```
